```
import math
```

```
#Menu de operaciones
while 1: 
    print("CALCULADORA")
    print("Debe elegir una opción")
    print("1-Suma")
    print("2-Resta")
    print("3-Multiplicación")
    print("4-División")
    print("5-División entera")
    print("6-Raiz Cuadrada")
    print("7-Factorial de un Número")
    print("8-Potenciacion al cuadrado")
    print("9-Potenciacion al cubo")
    print("10-Modulo")
    x= input("Escribe tu elección:\n")
    y=float(x)
  
    # Suma
    if(y==1 or y=='suma'):
        a=float(input("Ingresa un número\n"))
        b=float(input("Ingresa otro número\n"))
        print("La suma de",a,"+",b,'=',a+b)
        ejecutar = input("¿Desea ejecutar otra operando? (si/no)\n")
        if ejecutar== "no":
            break

    #resta
    elif(y==2):
        a=float(input("Ingresa un número\n"))
        b=float(input("Ingresa otro número\n"))
        print("La resta de:",a,"-",b,'=',a-b)
        ejecutar = input("¿Desea ejecutar otra operando? (si/no)\n")
        if ejecutar== "no":
            break
    
    #Multiplicación         
    elif(y==3):
        a=float(input("Ingresa un número\n"))
        b=float(input("Ingresa otro número\n"))
        print("La multiplicación de:",a,"*",b,'=',a*b)
        ejecutar = input("¿Desea ejecutar otra operando? (si/no)\n")
        if ejecutar== "no":
            break
    #Division
    elif(y==4):
        a=float(input("Ingresa un número\n"))
        b=float(input("Ingresa otro número\n"))
        if (b==0):
            print("No se puede dividir entre 0") 
        else:
            print("La división de:",a,"/",b,'=',a/b)
            ejecutar = input("¿Desea ejecutar otra operando? (si/no)\n")
            if ejecutar== "no":
              break
    #Division entera
    elif(y==5):
        a=float(input("Ingresa un número\n"))
        b=float(input("Ingresa otro número\n"))
        if (b==0):
            print("No se puede dividir entre 0") 
        else:
            print("La división entera de:",a,"/",b,'=',a//b)
            ejecutar = input("¿Desea ejecutar otra operando? (si/no)\n")
            if ejecutar== "no":
              break
    #Raiz Cuadrada
    elif(y==6):
        print("Ingresa un Numero")
        Raiz= float(input())
        print("La Raiz Cuadrada de:",Raiz,'es',math.sqrt(Raiz))
        ejecutar = input("¿Desea ejecutar otra operando? (si/no)\n")
        if ejecutar== "no":
            break
    #Factorial de un Numero
    elif(y==7):
        print("Ingrese un numero")
        valor= float(input())
        print("El factorial de",valor,"es: ",math.factorial(valor))
        ejecutar = input("¿Desea ejecutar otra operando? (si/no)\n")
        if ejecutar== "no":
            break
    #potencia cuadrada 
    elif(y==8):
        print("Ingrese un numero")
        valor= float(input())
        print("La potecia cuadrada de",valor,"es: ",valor**2)
        ejecutar = input("¿Desea ejecutar otra operando? (si/no)\n")
        if ejecutar== "no":
            break
    #potencia cubica 
    elif(y==9):
        print("Ingrese un numero")
        valor= float(input())
        print("La potecia cubica de",valor,"es: ",valor**3)
        ejecutar = input("¿Desea ejecutar otra operando? (si/no)\n")
        if ejecutar== "no":
            break
    #Modulo
    elif(y==10):
        a=float(input("Ingresa un número\n"))
        b=float(input("Ingresa otro número\n"))
        if (b==0):
            print("No se puede dividir entre 0") 
        else:
            print("La división de:",a,"/",b,'=',a%b)
            ejecutar = input("¿Desea ejecutar otra operando? (si/no)\n")
            if ejecutar== "no":
              break
else:
    print("Introduzca un opcion correcta ")
```
