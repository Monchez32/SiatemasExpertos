Como estrategia de fidelización de clientes, para la festividad de la natividad del 2020 la compañía va a regalar a cada cliente la selección si ésta tiene un costo menor a $25. Se solicita que codifique una solución para determinar el monto que la compañía va a invernir en esta campaña

1. Modificar el ejercicio 3 incluyendo el 13% en concepto del IVA de manera que el costo de la promocion sea debidamente cuantificada. Por medio del codigo fuente compare el costo sin IVA vs el nuevo costo e informe sobre la diferencia.

```
import time
import numpy as np 
```

```
listado_clientes = np.genfromtxt('costos.txt', delimiter='\n')
```

```
inicio = time.time()

costos_campaña = listado_clientes[listado_clientes < 25]
resultado_inversion = np.sum(costos_campaña)
ganancia_campaña = listado_clientes[listado_clientes >=25]
resultado_ganancia = np.sum(ganancia_campaña)

#calculo del  IVA del 13%
iva= 0.13
iva_inversion= resultado_inversion*iva
resultado_inversion_iva= resultado_inversion + iva_inversion
if (resultado_inversion_iva == resultado_inversion):
  print('tienen la misma cantidad de inversion ')
else:
  print('La diferencia es el 13% de IVA ')

print('Inversion en la campaña:','$',resultado_inversion)
print('IVA aplicarse es de:','$',iva_inversion)
print('Inversion con iva:','$',resultado_inversion_iva)
print('Ganancias obtenidas :','$',resultado_ganancia)
print('Duracion: {} segundos'.format(time.time() - inicio))
```
